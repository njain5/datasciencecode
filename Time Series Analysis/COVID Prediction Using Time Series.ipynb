Since, we are only worried about calculating number of cases, So we can ignore null values in other columns. Here we do not have any null values for Num Cases.

```python
data.describe()
```

For each date we need to sum up the Num cases and this will give us the time series data.

```python
data.groupby('Detected State')['Num cases'].sum()
```

**These were the number of cases on 25th April**

```python
ts = data.groupby('Date Announced')['Num cases'].sum()
ts
```

```python
df = pd.DataFrame(ts, index=None)
df.index = pd.to_datetime(df.index, dayfirst=True)
df.sort_index(inplace = True)
df
```

```python
df.head()
```

Here we sorting by the date.

```python
#Calculating the cumulative sum of covid cases for the day.

df['total cases'] = np.cumsum(df['Num cases'])
df
```

```python
df = df.drop("Num cases", axis =1)
```

# Auto ARIMA

<b>Akaike Information Criterion</b> - the lower the better the model <br>
AIC = n log(sse/n) + 2K

```python
!pip install pmdarima
```

```python
from pmdarima.arima import auto_arima
stepwise_model = auto_arima(df, start_p=1, start_q=1,
                           max_p=3, max_q=3, m=12,
                           start_P=0, seasonal=True,
                           d=1, D=1, trace=True,
                           error_action='ignore',  
                           suppress_warnings=True, 
                           stepwise=True)
print(stepwise_model.aic())
```

```python
# Making prediction for the next 30 days from 26 April to 26 May
future_forecast = stepwise_model.predict(n_periods=30)
print(future_forecast)
```

**According to the model the number of covid cases should be close to 1 lakh on 26th May. But the actual number of cases has reached to more than 1.5 lakh. The number of cases have increased drastically since relaxation of Lockdown 4.**
