```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Load the Data

```python
startup_data  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/startupdata/training_set_label.csv" )
```

```python
startup_data.head()
```

```python
startup_data.describe()
```

```python
startup_data.info()
```

**Note: There are few columns which are not required in order to predict the success of any startup. We can consider it later on given the time frame, I decide to drop those columns.**

```python
startup_data = startup_data.drop(['Unnamed: 0', 'latitude', 'longitude', 'zip_code', 'id', 'Unnamed: 6', 'city', 'name', 'founded_at', 
                                'closed_at', 'first_funding_at', 'last_funding_at', 'age_first_funding_year', 'age_last_funding_year',
                                'age_first_milestone_year', 'age_last_milestone_year', 'relationships', 'funding_rounds', 'milestones',
                                'state_code.1', 'state_code', 'category_code','object_id'],axis=1
                                )
```

```python
startup_data.info()
```

```python
startup_data.head()
```

```python
startup_data.isnull().sum()
```

###There are no missing values in the data now.

```python
X = startup_data.drop('status',1)
y = startup_data['status']
```

```python
new_y = []
for item in y:
    if item == 'closed':
      new_y.append(0)
    else:
      new_y.append(1)
```

```python
new_y = np.array(new_y)
```

```python
new_y
```

```python
X.info()
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)

#Scale and centre the data

X_normalized = scaler.transform(X)

#Create a pandas dataframe

X_normalized = pd.DataFrame(data = X, index= X.index, columns= X.columns)
```

### Split the data into Train and Test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_normalized, new_y, random_state=1234, test_size=0.3)
```

```python
#Building the Logistic Regression Model

from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(C=1.0, max_iter = 10000)  #Set large C value for low regularization to avoid overfitting
logreg.fit(X_train,y_train)

print(logreg.coef_)
print(logreg.intercept_)
```

```python
#Now applying the learnt model on test and train data as well.

y_pred_test = logreg.predict(X_test)
y_pred_train = logreg.predict(X_train)
```

```python
from sklearn import metrics

#comparing the metrics of predicted label and real label of test data

print('Test Accuracy: ', metrics.accuracy_score(y_test, y_pred_test))
```

```python
# comparing the metrics of predicted lebel and real label of test data
print("Train Accuracy: ", metrics.accuracy_score(y_train, y_pred_train))
```

```python
#Confusion Matrix

conf = metrics.confusion_matrix(y_test, y_pred_test)
```

```python
cmap = sns.cubehelix_palette(50, hue=0.05, rot=0, light=0.9, dark=0, as_cmap=True)
sns.heatmap(conf,cmap = cmap, xticklabels = ['Prediction No', 'Prediction Yes'], yticklabels=['Actual No', 'Actual yes'], 
            annot=True, fmt ='d')
```

```python
#Creating Classification Report

cr = metrics.classification_report(y_test, y_pred_test)
print(cr)
```

```python
# Importing Decision Tree and performing decision tree
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(X_train,y_train)
dt_pred = dt.predict(X_test)
```

```python
# RMSE score for Decision Tree
dt_rmse = np.sqrt(metrics.mean_squared_error(dt_pred,y_test))
dt_rmse
```

```python
# R2 score for Decision Tree
dt_r2score = metrics.r2_score(dt_pred,y_test)
dt_r2score
```

```python
# Train score for Decision Tree
dt_train = dt.score(X_train,y_train)
dt_train
```

```python
# Test score for Decision Tree
dt_test = dt.score(X_test,y_test)
dt_test
```
