### Patient Survival in a Hospital

### Objective

A hospital in the province of Greenland has been trying to improve its care conditions by looking at historic survival of the patients. They tried looking at their data but could not identify the main factors leading to high survivals.<br>

You are the best data scientist in Greenland and they've hired you to solve this problem. Now you are responsible for developing a model that will predict the chances of survival of a patient after 1 year of treatment (Survived_1_year).

### Dataset & Data Description

The dataset contains the patient records collected from a hospital in Greenland. The "Survived_1_year" column is a target variable which has binary entries (0 or 1).<br>
•	Survived_1_year == 0, implies that the patient did not survive after 1 year of treatment<br>
•	Survived_1_year == 1, implies that the patient survived after 1 year of treatment

Data Description:<br>
•	ID_Patient_Care_Situation: Care situation of a patient during treatment<br>
•	Diagnosed_Condition: The diagnosed condition of the patient<br>
•	ID_Patient: Patient identifier number<br>
•	Treatment_with_drugs: Class of drugs used during treatment<br>
•	Survived_1_year: If the patient survived after one year (0 means did not survive; 1 means survived)<br>
•	Patient_Age: Age of the patient<br>
•	Patient_Body_Mass_Index: A calculated value based on the patient’s weight, height, etc.<br>
•	Patient_Smoker: If the patient was a smoker or not<br>
•	Patient_Rural_Urban: If the patient stayed in Rural or Urban part of the country<br>
•	Previous_Condition: Condition of the patient before the start of the treatment ( This variable is splitted into 8 columns - A, B, C, D, E, F, Z and Number_of_prev_cond. A, B, C, D, E, F and Z are the previous conditions of the patient. Suppose for one patient, if the entry in column A is 1, it means that the previous condition of the patient was A. If the patient didn't have that condition, it is 0 and same for other conditions. If a patient has previous condition as A and C , columns A and C will have entries as 1 and 1 respectively while the other column B, D, E, F, Z will have entries 0, 0, 0, 0, 0 respectively. The column Number_of_prev_cond will have entry as 2 i.e. 1 + 0 + 1 + 0 + 0 + 0 + 0 + 0 = 2 in this case. )

### Observations and Resolution

From above we can clearly say that there are null values for Number_of_prev_cond column and A,B,C,D,E,F,Z. We can drop these values, and I will also drop Number_of_prev_cond column since it add up all the previous conditions in A,B,C,D,E,F,Z.<br>

We will convert Treated_with_drugs, Patient_Smoker, Patient_Rural_Urban and Patient_mental_condition into categorical variables. 

```python
pharma_data.isna().sum()
```

```python
#Removing all the null values.

pharma_data.dropna(inplace=True)
```

```python
pharma_data.isna().sum()
```

**Note: All the null values have been removed.**

```python
#Dropping Number_of_prev_cond column.

pharma_data = pharma_data.drop(['Number_of_prev_cond'], axis = 1)
```

```python
pharma_data.head()
```

**Column Number_of_prev_cond has been dropped.**

```python
#Converting columns to categorical one.

pharma_data = pd.concat([pharma_data,pd.get_dummies(pharma_data['Treated_with_drugs'],prefix='Treated_with_drugs')], axis=1)
del pharma_data['Treated_with_drugs']
```

```python
pharma_data.head()
```

```python
pharma_data = pd.concat([pharma_data,pd.get_dummies(pharma_data['Patient_Smoker'],prefix='Patient_Smoker')], axis=1)
del pharma_data['Patient_Smoker']
```

```python
pharma_data = pd.concat([pharma_data,pd.get_dummies(pharma_data['Patient_Rural_Urban'],prefix='Patient_Rural_Urban')], axis=1)
del pharma_data['Patient_Rural_Urban']
```

```python
pharma_data = pd.concat([pharma_data,pd.get_dummies(pharma_data['Patient_mental_condition'],prefix='Patient_mental_condition')], axis=1)
del pharma_data['Patient_mental_condition']
```

```python
pharma_data.head()
```

```python
pharma_data.info()
```

NOTE: Since All the patients are stable, We can drop Mental condition column.

```python
#Dropping Patient_mental_condition_Stable.

pharma_data = pharma_data.drop(['Patient_mental_condition_Stable'], axis = 1)
```

```python
#Standardize rows into uniform scale.

X= pharma_data.drop(['Survived_1_year'], axis=1)
y= pharma_data['Survived_1_year']

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)

#Scale and centre the data

pharma_data_normalized = scaler.transform(X)

#Create a pandas dataframe

pharma_data_normalized = pd.DataFrame(data = X, index= X.index, columns= X.columns)
```

```python
pharma_data_normalized.info()
```

### Split the data into train and test.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(pharma_data_normalized, y, random_state=1234, test_size=0.3)
```

```python
#Building the Logistic Regression Model

from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(C=1.0, max_iter = 10000)  #Set large C value for low regularization to avoid overfitting
logreg.fit(X_train,y_train)

print(logreg.coef_)
print(logreg.intercept_)
```

```python
#Now applying the learnt model on test and train data as well.

y_pred_test = logreg.predict(X_test)
y_pred_train = logreg.predict(X_train)
```

### Metrics of the model

```python
from sklearn import metrics

#comparing the metrics of predicted label and real label of test data

print('Test Accuracy: ', metrics.accuracy_score(y_test, y_pred_test))
```

```python
# comparing the metrics of predicted lebel and real label of test data
print("Train Accuracy: ", metrics.accuracy_score(y_train, y_pred_train))
```

```python
#Null Accuracy

#Actual value of y_test

y_test.value_counts()
y_test.value_counts().head(1)/ len(y_test)
```

So, model performance is better than null accuracy.

```python
#Confusion Matrix

conf = metrics.confusion_matrix(y_test, y_pred_test)
```

```python
cmap = sns.cubehelix_palette(50, hue=0.05, rot=0, light=0.9, dark=0, as_cmap=True)
sns.heatmap(conf,cmap = cmap, xticklabels = ['Prediction No', 'Prediction Yes'], yticklabels=['Actual No', 'Actual yes'], 
            annot=True, fmt ='d')
```

```python
#Creating Classification Report

cr = metrics.classification_report(y_test, y_pred_test)
print(cr)
```
